```python
from spacepy import pycdf
import pandas as pd
from datetime import datetime
import numpy as np
import datetime as dt
```

```python
cdf = pycdf.CDF('ac_h0s_mfi_20120122000004_20120122235948_cdaweb.cdf')
print(cdf['Epoch'])
print(cdf['Epoch'][...])
```

```python
ace_mfi = pd.read_csv('AC_H0_MFI_3649384.csv', header=0, parse_dates=[0], names=['Date', 'Bx', 'By', 'Bz', 'X', 'Y', 'Z'])
B = ace_mfi[['Bx', 'By', 'Bz']].to_numpy()
```

```python
cdf_ace = pycdf.CDF('ac_mfi_fromcsv1.cdf','')
tt2000_datetimes = ace_mfi['Date'][...]
epoch_values = [dt.timestamp() for dt in tt2000_datetimes]
print(epoch_values)
cdf_ace['Epoch'] = epoch_values
cdf_ace['B'] = B.astype(np.float32)
cdf_ace.attrs['Author'] = 'Espen Fredrick'
cdf_ace.attrs['CreateDate'] = datetime.now()
cdf_ace['B'].attrs['units'] = 'nT'
cdf_ace.close()
```
